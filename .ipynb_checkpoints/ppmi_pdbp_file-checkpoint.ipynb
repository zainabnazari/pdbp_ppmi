# The table for RNA-Seq from PPMI and PDBP
The tables are seperated from their diagnosis and sex values, in this part we combine them accordingly.

```python
import pandas as pd
import numpy as np
from pathlib import Path
from datetime import datetime
```

```python
path1=Path("/Users/zainabnazari/Desktop/pdbp/PPMI_PDBP_Log2CPM_sex_effect_removed_Norm_Quantile_together_PD_CTR")
```

# PPMI 

```python
pdbp_ppmi_data1=pd.read_csv(path1/"PPMI_subset_from_PPMI_PDBP_shared_genes_Log2CPM_sex_effect_removed_QuantileNorm_02giu2024.txt",delimiter='\t')
t_ppmi = pdbp_ppmi_data1.T
```

```python
t_ppmi.reset_index(inplace=True)
```

```python
t_ppmi['index']= t_ppmi['index'].str[1:]
```

```python
t_ppmi.rename(columns={'index':'ID'}, inplace=True)
```

```python
t_ppmi.set_index('ID', inplace=True)
```

```python
t_ppmi
```

```python
factor_ppmi=pd.read_csv(path1/"factor_PPMI_subset_from_PPMI_PDBP_shared_genes_Log2CPM_sex_effect_removed_QuantileNorm_02giu2024.txt",delimiter='\t')
factor_ppmi['ID'] = factor_ppmi['ID'].astype(str)
```

```python
combined_ppmi = pd.merge(t_ppmi,factor_ppmi, on='ID')
combined_ppmi
```

```python
combined_ppmi.to_csv(path1/'ppmi_data.csv', sep='\t', index=False)
```

# PDBP

```python
pdbp_ppmi_data=pd.read_csv(path1/"PDBP_subset_from_PPMI_PDBP_shared_genes_Log2CPM_sex_effect_removed_QuantileNorm_02giu2024.txt",delimiter='\t')
t_pdbp = pdbp_ppmi_data.T
```

```python
index_pdbp = t_pdbp.reset_index()
```

```python
index_pdbp['index']= index_pdbp['index'].str[3:]
```

```python
index_pdbp
```

```python
index_pdbp.rename(columns={"index":"ID"},inplace=True)
```

```python
index_pdbp.set_index('ID', inplace=True)
```

```python
index_pdbp
```

```python
factor_pdbp=pd.read_csv(path1/"factor_PDBP_subset_from_PPMI_PDBP_shared_genes_Log2CPM_sex_effect_removed_QuantileNorm_02giu2024.txt",delimiter='\t')
factor_pdbp
```

```python
factor_pdbp['ID'] = factor_pdbp['ID'].str[3:]
factor_pdbp
```

```python
combined_pdbp = pd.merge(index_pdbp, factor_pdbp, on="ID")
```

```python
combined_pdbp
```

```python
combined_pdbp.to_csv(path1/"pdbp_data.csv",sep='\t', index=False)
```
